```python
import pandas as pd
```

```python
dataset=pd.read_csv("file_name.csv")
x=dataset.iloc[:,:-1].values #choose acc to your data and purpose
y=dataset.iloc[:,-1].values # choose acc to your data and purpose
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(x_train)
sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2) # for euclidean distance
knc.fit(x_train,y_train)
```

```python
y_pred=knc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
print(score)
```
